```python
import tensorflow as tf
tf.__version__
```

```python
import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
celsius_q    = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
fahrenheit_a = np.array([-40,  14, 32, 46.4, 59, 71.6, 100.4],  dtype=float)

for i,c in enumerate(celsius_q):
  print("{} degrees Celsius = {} degrees Fahrenheit".format(c, fahrenheit_a[i]))
```

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```python
model = tf.keras.Sequential([l0])
```

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celsius_q, fahrenheit_a, epochs=1000, verbose=True)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
plt.show()
```

```python
print(model.predict([100.0]))
```

```python
print("These are the layer variables: {}".format(l0.get_weights()))
```

```python
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=4)
l3 = tf.keras.layers.Dense(units=1)
```

```python
model = tf.keras.Sequential([l0 , l1, l2 ,l3])
```

```python
model.compile(loss=tf.losses.mean_squared_error , optimizer=tf.keras.optimizers.Adam(0.01))
```

```python
history = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
plt.show()
```

```python
print(model.predict([100.0]))
```
